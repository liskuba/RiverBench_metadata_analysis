```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
directory = "../datasets/dev#statistics-full"
metadata = "https://w3id.org/riverbench/schema/metadata#SimpleLiteralCountStatistics"
files = [file for file in os.listdir(directory)]
names, means = [], []
for file in files:
    df = pd.read_csv(f"{directory}/{file}")
    mean = df.loc[df["metadata"] == metadata, "mean"].iloc[0]
    names.append(file[:-4])
    means.append(mean) 
```

```python
plt.barh(names, means)
plt.show()
```
